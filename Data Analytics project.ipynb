# AI's Impact on Student Performance and Wellbeing

#### Problem Statement:

Generative AI tools have become a regular part of student life, but it's still unclear whether they're actually helping students learn or just helping them get things done faster. This project looks at data from 50,000 students to understand the relationship between how much they use AI tools, how skilled they are at using them, and the policies their institutions have in place — and how all of that connects to changes in their GPA, how much they feel they depend on AI, their anxiety during exams, and burnout. The aim is to figure out which usage patterns actually support learning and which ones might be signs of over-reliance, so the findings can help shape better AI-literacy guidance and institutional policies going forward.

### Stage 1: Collect

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("ai_student_impact_dataset (1).csv")
print("Rows, Columns:",df.shape)
df.head()
```

```python
df.info()
```

### Stage 2: Clean

```python
print("Missing values in each column:")
print(df.isnull().sum())
print()
print("Duplicate rows:", df.duplicated().sum())
print()
print("Major_Category column has these values:", df["Major_Category"].unique())
```

```python
# Fix 1: remove duplicate rows
print("Before:", df.shape)
df = df.drop_duplicates()
print("After:", df.shape)
```

### Stage 3: Explore

```python
print(f"Average Pre-Semester GPA: {df['Pre_Semester_GPA'].mean():.2f}")
print(f"Average Post-Semester GPA: {df['Post_Semester_GPA'].mean():.2f}")
print(f"Average Weekly GenAI Hours: {df['Weekly_GenAI_Hours'].mean():.2f}")
print(f"Average Perceived AI Dependency: {df['Perceived_AI_Dependency'].mean():.2f}")
print(f"Average Skill Retention Score: {df['Skill_Retention_Score'].mean():.2f}/100")
print("Average GPA Improvement:", round((df["Post_Semester_GPA"] - df["Pre_Semester_GPA"]).mean(), 2))
```

### Stage 4: Visualize

#### Histogram

```python
plt.figure(figsize=(7, 5))
#df["GPA_Change"] = df["Post_Semester_GPA"] - df["Pre_Semester_GPA"]
plt.hist(df["GPA_Change"], bins=7, color="orange", edgecolor="black") 
plt.suptitle("Distribution of GPA Change", fontsize=14, fontweight="bold")
plt.title("Difference between Post-Semester GPA and Pre-Semester GPA", fontsize=10, color="gray")
plt.xlabel("GPA_Change")
plt.ylabel("Count")
plt.show()
```

##### Analysis based on the histogram
While the majority of students (87.5%) showed GPA improvement over the semester (mean change ≈ +0.20), a notable minority (~12%) experienced a decline, with some drops as large as -0.92. This suggests GenAI usage is not uniformly beneficial, and warrants deeper analysis into which student profiles fall into the declining group.

#### Scatter plot

```python
plt.figure(figsize=(7, 5))
plt.scatter(df["Weekly_GenAI_Hours"], df["GPA_Change"], color="teal",s=0.5,alpha=0.1)
plt.title("Weekly GenAI Usage Hours vs GPA Change")
plt.xlabel("Weekly_GenAI_Hours")
plt.ylabel("GPA_Change")
plt.show()
```

##### Analysis based on the Scatter plot
Although the linear correlation between weekly GenAI usage and GPA change is negligible (r ≈ -0.05), binning students by usage intensity reveals a non-linear relationship: GPA improvement peaks for moderate users (5–15 hrs/week, ~+0.23) and declines for heavy users (35–40 hrs/week, ~+0.12), suggesting diminishing or even reversing returns beyond moderate engagement.

#### Bar graph

```python
category_sales = df.groupby("Major_Category")["GPA_Change"].mean()
plt.figure(figsize=(7, 4))
plt.bar(category_sales.index, category_sales.values, color=["#4C72B0", "#55A868", "#C44E52", "#8172B2", "#CCB974"])
plt.title("Average GPA Change by Major Category")
plt.xlabel("Major_Category")
plt.ylabel("Avg GPA Change")
plt.show()
```

##### Analysis based on the bar graph
STEM students showed the highest average GPA improvement (+0.22) compared to other majors (range +0.19 to +0.20), a difference confirmed as statistically significant (ANOVA p < 0.0001) despite the modest absolute size — likely a reflection of large sample size rather than a strong practical effect. STEM students also displayed greater variance in outcomes, suggesting more divergent individual experiences with GenAI tools.

#### Box Plot (Modification to be done- for outliers)

```python
avg_retention_by_usecase = df.groupby("Primary_Use_Case")["Skill_Retention_Score"].mean()
plt.figure(figsize=(7, 5))
plt.bar(avg_retention_by_usecase.index, avg_retention_by_usecase.values, color="skyblue", edgecolor="black")
plt.title("Primary Use Case affecting Skill Retention")
plt.xlabel("Primary_Use_Case")
plt.ylabel("Avg Skill_Retention_Score")
plt.xticks(rotation=45, ha="right")
plt.show()
```

##### Analysis based on the box plot
Average skill retention varied meaningfully by primary AI use case (ANOVA, F = 140.9, p < 0.0001), ranging from 78.1 for debugging/troubleshooting to 73.7 for direct answer generation. While the difference in means is moderate, examining the full distribution reveals direct-answer users also show the greatest variability and the most extreme low-retention cases, reinforcing this use case as the highest-risk pattern for skill erosion.
